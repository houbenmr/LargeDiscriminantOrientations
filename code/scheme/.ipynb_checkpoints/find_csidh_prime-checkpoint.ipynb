```python
from scheme_cost import *

from math import log
```

```python
# We find a prime number of the form 4 * prod_i ell_i - 1 of suitable size

B = ell = 3

ells = [3]

while log(B, 2) <= 200:
    ell = ell.next_prime()
    ells.append(ell)
    B *= ell
    
print('the largest ell_i is', max(ells))
print('the number of ell_i is', len(ells))
    
for ell1 in ells:
    for ell2 in ells:
        ells_without_ell12 = [x for x in ells if not x == ell1 and not x == ell2]
        M = 4*prod(ells_without_ell12)
        p = M-1
        if p.is_pseudoprime():
            print(ell1, ell2, prime_score(p, 221))
```
